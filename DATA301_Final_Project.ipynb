````
import urllib
import time

def get_summoner_puuid(name):
  encoded_name = urllib.parse.quote(name)
  response = requests.get("https://na1.api.riotgames.com/tft/summoner/v1/summoners/by-name/" + encoded_name + "?api_key=" + temp_key)
  data_summoner = json.loads(response.text)
  df_summoner = json_normalize(data_summoner)
  time.sleep(0.5)
  # puuid
  return df_summoner.iloc[0][2]

df_challenger["puuid"] = df_challenger["summonerName"].map(get_summoner_puuid)
````

````
df_challenger
````

````
def get_matches_data(puuid):
  response = requests.get("https://americas.api.riotgames.com/tft/match/v1/matches/by-puuid/" + puuid + "/ids?count=5&api_key=" + temp_key)
  matches = json.loads(response.text)
  match_data = []
  for match in matches:
    response = requests.get("https://americas.api.riotgames.com/tft/match/v1/matches/" + match + "?api_key=" + temp_key)
    match_data.append(json.loads(response.text))
    time.sleep(1.5)
  return json.dumps(match_data)

df_challenger["matches"] = df_challenger["puuid"].map(get_matches_data)
````

````
total_traits_dict = {}

def eval_part_traits(participant, traits_dict):
  for trait in participant["traits"]:
    if trait["tier_current"] > 0:
      if trait["name"] not in traits_dict:
        traits_dict[trait["name"]] = 1
      else:
        traits_dict[trait["name"]] += 1
  
def eval_match_traits(participants, traits_dict):
  participants_json = json.loads(participants.replace("'", "\""))
  for participant in participants_json:
    eval_part_traits(participant, traits_dict)

df_matches["info.participants"].map(lambda x : eval_match_traits(x, total_traits_dict))
````

### Machine Learning Model

````
total_trait_names = list(df_traits.index)
````

````
def get_part_traits(participant):
  traits = {trait_name: 0 for trait_name in total_trait_names}

  for trait in participant["traits"]:
    if trait["tier_current"] > 0:
      traits[trait["name"]] = 1
  return traits
  
def combine_match_traits(participants, match_traits_list):
  participants_json = json.loads(participants.replace("'", "\""))
  for participant in participants_json:
    match_traits_list.append(get_part_traits(participant))

match_traits_list = []
df_matches["info.participants"].map(lambda x : combine_match_traits(x, match_traits_list))
````

````
df_match_traits = pd.DataFrame(match_traits_list)
df_match_traits
````

#### Implementing a K-means clustering model to predict overall team compositions

````
import numpy as np
````

````
from sklearn.cluster import KMeans

model = KMeans(n_clusters=20)
model.fit(df_match_traits)

centroids = model.cluster_centers_
clusters = model.labels_

print(clusters)
````

````
def generate_traits_input(desired_traits):
  traits = {trait_name: 0.3 for trait_name in total_trait_names}
  for trait in desired_traits:
    traits[trait] = 1
  return list(traits.values())

def convert_input_to_traits(trait_input):
  trait_names = []
  for idx, input in enumerate(trait_input):
    if input:
      trait_names.append(total_trait_names[idx])
  return trait_names
````

````
desired_traits = ["Set4_Executioner", "Divine"]
predicted_cluster = model.predict(
    [generate_traits_input(desired_traits)]
)[0]
predicted_centroid = centroids[predicted_cluster]
trait_input = np.rint(predicted_centroid)
predicted_traits = convert_input_to_traits(trait_input)

print("{} -> {}".format(desired_traits, predicted_traits))
````

````
prediction_map = {}
for trait in total_trait_names:
  desired_traits = [trait]
  predicted_cluster = model.predict(
      [generate_traits_input(desired_traits)]
  )[0]
  predicted_centroid = centroids[predicted_cluster]
  trait_input = np.rint(predicted_centroid)
  predicted_traits = convert_input_to_traits(trait_input)

  prediction_map[desired_traits[0]] = predicted_traits
````
